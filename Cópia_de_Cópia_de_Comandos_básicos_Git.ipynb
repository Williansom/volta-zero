# **Demonstração inicial de como utilizar a biblioteca OS do Python**

Antes de nos aprofundarmos em cenários mais complexos de uso do Git, é fundamental entender os conceitos e comandos básicos. Nesta seção  você será guiado por uma série de passos que demonstram o uso fundamental do Git.

**Inicialmente, é preciso criar uma função para executar comandos de terminal em Python e imprimir sua saída no Goolge Colab.**

```python
# Importando as bibliotecas do Python necessárias
import os
import subprocess

# Função para executar comandos de terminal em Python e imprimir sua saída no Goolge Colab
def execute_command(cmd):
    print(f"Executing: {cmd}")
    process = subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    stdout, stderr = process.communicate()
    print(stdout.decode())
    print(stderr.decode())
    print("-" * 50)

a = execute_command("listdir")
print(a)
```

A biblioteca os em Python é uma biblioteca padrão que fornece uma interface para interagir com funcionalidades dependentes do sistema operacional. Ela permite que um programa Python acesse várias funcionalidades do sistema operacional, como manipulação de arquivos, leitura de variáveis de ambiente, execução de comandos do sistema, entre outras.

#Criando uma branch.

Agora que você conheceu como realizar os comandos básicos do git através do Python, vamos praticar um pouco como trabalhar com branches. Nesta parte, você deverá realizar as seguintes etapas:

> Crie uma nova branch chamada "feature-1".

> Adicione no arquivo data.txt números do 21 até o 30.

> Adicione estas alterações do arquivo data.txt na branch.

> Realize o o commit destas alterações com a mensagem "Adicionados números 21 até 31 para a feature-1".

> Altere para a branch "master".

> Realize o merge da branch "feature-1" com a branch "master".

> Verifique se o contéudo foi alterado corretamente na branch "master".

**6. Branching:**

Em Git, "branches" ou ramificações são uma maneira de você experimentar novas ideias ou recursos sem afetar a linha principal de desenvolvimento.

Ao usar git checkout -b [nome-da-branch], você está criando e alternando para uma nova branch.

Lembre-se, as alterações feitas em uma branch não afetarão outras branches até que você decida integrá-las.

```python
# Crie e ative a branch feature-1.
execute_command("git checkout -b feature-1")

# Adicione os números 20 até 31 no arquivo data.txt
with open("data.txt", "a") as f:
    for i in range(21, 31):
        f.write(str(i) + '\n')
# Adicione estas alterações do arquivo data.txt na branch.
execute_command("git add data.txt")
# Realize o o commit destas alterações com a mensagem "Adicionados números 21 até 31 para a feature-1".
execute_command('git commit -m "Adicionados números 21 até 31 para a feature-1"')
```

**7. Alternando entre Branches:**

Você pode facilmente alternar entre branches usando o comando git checkout. Isso é útil para retornar à linha principal de desenvolvimento ou para revisitar uma feature específica.

```python
# Altere para a branch master
execute_command("git checkout master")
```

**8. Realizando o merge entre branches:**

O processo de trazer alterações de uma branch para outra é chamado de "merge".
Ao executar um git merge [nome-da-branch], você está integrando as alterações dessa branch no branch atual.

```python
# Realize o merge da branch "feature-1" com a branch "master".
execute_command("git merge feature-1")

# Verifique se o contéudo foi alterado corretamente na branch "master".
execute_command("cat data.txt")
```

# **Resolvendo um conflito de merge**.

Uma vez que você está familiarizado com a realização dos comandos básicos do Git utilizando Python, agora será capaz de resolver um conflito de merge.

Um conflito de merge no Git ocorre quando duas ou mais pessoas editam a mesma parte de um arquivo e depois tentam combinar (ou "mergear") essas alterações em uma única linha de desenvolvimento. Como o Git não sabe automaticamente qual das edições deve prevalecer, ele marca o arquivo como "em conflito" e pede intervenção manual para resolver o problema.

Vamos entender com um exemplo:

Imagine que você e seu colega estão trabalhando em um projeto. Ambos decidem modificar a mesma linha de um arquivo. Você altera essa linha para dizer "Olá, Mundo!" e seu colega a altera para "Oi, Terra!". Ambos fazem commits com suas respectivas mudanças em branches separadas. Quando um de vocês tenta fazer um merge dessas branches, o Git detectará que as mesmas linhas foram alteradas em ambas as branches e não sabe qual delas deve usar. Isso é um conflito de merge.

Quando ocorre um conflito, o Git insere marcas especiais no arquivo conflitante para ajudá-lo a identificar onde está o problema:

Captura de Tela 2023-08-09 às 20.04.35.png

Neste exemplo:

A seção entre <<<<<<< HEAD e ======= mostra as alterações da branch atual (normalmente a branch "master" ou a branch em que você está tentando fazer o merge).

A seção entre ======= e >>>>>>> [branch-name] mostra as alterações da outra branch.

Para resolver o conflito:

> Abra o arquivo e decida qual das alterações deseja manter, ou crie uma nova alteração que combine as duas.

> Remova as marcas de conflito (<<<<<<<, =======, e >>>>>>>) do arquivo.

> Faça um commit das alterações para finalizar a resolução do conflito.

É importante entender que um conflito de merge não é necessariamente um "erro" ou um "problema" – é apenas uma situação em que o Git precisa de instruções claras do desenvolvedor sobre como proceder. Resolver conflitos é uma habilidade essencial para quem trabalha com Git em projetos colaborativos.

# **Simulando o conflito de merge**

```python
# Escrevendo sobre o evento fictício de 1990
with open("history.txt", "w") as f:
    f.write("In 1990, the first website was published.\n")
execute_command("git add history.txt")
execute_command('git commit -m "Added event from 1990"')

# Criando a branch alternative-history
execute_command("git checkout -b alternative-history")

# Alterando o evento de 1990 na branch alternative-history
with open("history.txt", "w") as f:
    f.write("In 1990, aliens visited Earth for the first time.\n")
execute_command("git add history.txt")
execute_command('git commit -m "Changed 1990 event in alternative history"')

# Voltando para a branch master
execute_command("git checkout master")

# Adicionando um evento de 1995 na branch master
with open("history.txt", "a") as f:
    f.write("In 1995, the Java programming language was released.\n")
execute_command("git add history.txt")
execute_command('git commit -m "Added event from 1995"')
```

**Git Rebase**

O rebase é uma das duas principais formas de integrar mudanças de uma branch para outra no Git (a outra é o merge). O rebase é um processo que pega as alterações de um branch e as "reaplica" em cima de outro. Isso pode parecer similar ao merge, mas o resultado e a abordagem são diferentes.

**Como funciona:**

Imagine que você tem uma base de código com uma branch principal (por exemplo, master) e uma branch de feature em que você está trabalhando. Durante o seu trabalho na branch de feature, a branch master recebe novos commits. Isso pode fazer com que a sua branch de feature fique "desatualizada".

Ao executar um rebase, em vez de simplesmente "juntar" as duas branches, o Git vai pegar todos os commits da sua branch de feature, "guardá-los de lado" temporariamente, e depois aplicá-los um a um em cima da branch master.

Vamos imaginar que o Git é como um livro de histórias do seu projeto e que as alterações no projeto são como capítulos desse livro. O git rebase é como rearranjar ou reescrever os capítulos do livro para torná-los mais organizados.

Em termos mais simples:

Git Merge (Mesclar): Adiciona um novo capítulo ao final do livro. As mudanças são integradas, mas a linha do tempo original permanece.

Git Rebase (Rebase): Pega os capítulos do livro que você escreveu, os guarda temporariamente, e então aplica os capítulos atualizados de outra pessoa. Depois, adiciona seus capítulos de volta no final. Isso reescreve a história de uma maneira mais linear e organizada.

**Comparação com Merge:**

> **Histórico Linear vs. Histórico Não Linear:** Uma das maiores diferenças entre rebase e merge é que o rebase cria um histórico linear, enquanto o merge pode resultar em um histórico não linear (com bifurcações e junções). Muitos desenvolvedores preferem um histórico linear porque pode ser mais fácil de seguir e entender.

> **Conflitos:** Com ambos, rebase e merge, você pode encontrar conflitos que precisam ser resolvidos manualmente. No entanto, no caso de um rebase, você resolverá conflitos para cada commit individualmente, enquanto com um merge, você resolverá todos os conflitos de uma só vez.

**Vantagens do Rebase:**

> **Histórico Limpo:** Como mencionado, o rebase ajuda a manter um histórico de commits mais linear e limpo.

> **Evita Commits de Merge:** Ao usar o rebase, você não terá commits adicionais no histórico indicando um merge, a menos que você faça um merge explícito posteriormente.

**Desvantagens:**

> **Histórico Reescrito:** O rebase efetivamente reescreve o histórico do commit. Isso pode ser problemático se mais pessoas estiverem trabalhando na mesma branch, pois o histórico alterado pode criar confusões.

> **Pode ser mais complexo para iniciantes:** Para aqueles que são novos no Git, o rebase pode parecer um pouco mais mágico e menos intuitivo do que um simples merge.

**Quando usar?**

Use rebase para integrar mudanças de uma branch para outra quando quiser manter um histórico linear.

Use merge quando quiser preservar o histórico completo e explícito de alterações, ou quando trabalhar em branches compartilhadas com outros desenvolvedores.

**Nota Importante:**

Nunca faça rebase em branches que estão sendo usadas por outras pessoas ou em branches públicas, como branches em repositórios remotos que outras pessoas estão usando. Isso porque o rebase reescreve o histórico e pode criar confusão para outros desenvolvedores que têm uma cópia do histórico anterior.

O rebase é uma ferramenta poderosa no arsenal do Git, mas, como todas as ferramentas poderosas, deve ser usada com compreensão e cautela.

```python
# Tentando realizar o rebase da alternative-history em master e gerando o conflito.
execute_command("git rebase alternative-history")
```

# **Resolvendo o conflito de merge**

```python
# Abra, edite e feche o arquivo "history.txt" para resolver o conflito.
file = open("history.txt", "w")
file.write("In 1990, aliens visited Earth for the first time.\n")  # Evento da branch alternative-history
file.write("In 1995, the Java programming language was released.\n")  # Evento da branch master
file.close()
```

```python
# Adicione o arquivo resolvido ao índice
execute_command("git add history.txt")
```
